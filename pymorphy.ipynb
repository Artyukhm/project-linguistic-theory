```python
import pandas as pd
import pymorphy2
from nltk.tokenize import word_tokenize
```

```python
morph = pymorphy2.MorphAnalyzer()
```

```python
df = pd.read_excel("Новая таблица.xlsx")

df.head()
```

```python
df.drop_duplicates()
```

```python
new_df_list = []

for id, sentence in enumerate(df["sentence"].values):
    tokens = word_tokenize(sentence)
    for token in tokens:
        parsed_token = morph.parse(token)[0]
        new_dict = {
            "sentence_id": id,
            "normal_form": parsed_token.normal_form,
            "pos": parsed_token.tag.POS,
            "gender": parsed_token.tag.gender,
            "case": parsed_token.tag.case,
            "person": parsed_token.tag.person,
            "tag": parsed_token.tag}
        new_df_list.append(new_dict)
```

```python
new_df = pd.DataFrame(new_df_list)

new_df.head()
```

```python
new_df.to_excel('pymorphy.xlsx', index=False)
```
